<h2>Higher-order methods: Runge-Kutta</h2>

```python
import numpy as np
from matplotlib import pyplot as plt
```

In the last section, we explored how to apply the Euler-Cromer (aka leapfrog) scheme to solve differential equations resulting from Newton's second law in mechanical systems. Those mechanical systems have a nice type of symmetry that lends them to the reasonably conservative Euler-Cromer method. Let's consider more general differential equations and introduce a fairly efficient method for numerical integration known as the <b>fourth-order Runge-Kutta (RK4)</b> scheme.

<h2>Radioactive decay</h2>

A particular sample of some radioactive isotope begins with some amount $A(0)$ at time $t=0$. The quantitative principle governing this process is

$$\frac{dA}{dt} = -\lambda A,$$

where $\lambda$ is the decay constant with dimensions of inverse time. That is, the rate of decay is proportional to the amount present. This model is one of the simplest possible differential equations, so an exact solution is available:

$$A(t) = A(0) e^{-\lambda t}$$


<h3>Forward-Euler</h3>

Like the harmonic oscillator, this model gives us a handy benchmark for the approximate numerical procedures that we'll apply to differential equations. We have already mentioned the <b>forward-Euler</b> approach which is to take

$$\frac{dA}{dt} \rightarrow \frac{\Delta A}{\Delta t}$$ 

where $\Delta A = A(t+\Delta t) - A(t)$. The right-hand side is then replaced by $-\lambda A(t)$ so that

$$A(t+\Delta t) \approx A(t) + \Delta t(-\lambda A(t))$$

Let's take $A(0) = 1000$ (of some units), $\lambda = 0.5 \mbox{min}^{-1}$ and attempt to work out $A(t)$ numerically using forward Euler.

```python
N = 1000
tmax = 10
t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

λ = 0.5
Ao = 1000
A = np.zeros(N)
A[0] = Ao

for i in range(0,N-1):
    A[i+1] = A[i] + dt*(-λ*A[i])

plt.plot(t,np.abs(Ao*np.exp(-λ*t)-A))
#plt.plot(t,Ao*np.exp(-k*t))
plt.xlabel('t')
plt.ylabel('A(t)')
plt.show()
```

At least the error doesn't blow up at long times. This will not always be the case. In what follows, we introduce two methods for actually improving the error with the same step size $\Delta t$. Recall that forward-Euler creates $\mathcal{O}(\Delta t)$ error. Our first modified method will reduce errors to $\mathcal{O}(\Delta t^{2})$.

As a side note (in case you were wondering), there <i>is</i> a method known as <b>backwards Euler</b>. The crucial change in this setting is to replace $N$ on the right-hand side with $A(t+\Delta t)$ instead of $A(t)$. This method turns out to give the same level of error, but it leads to much greater numerical <i>stability</i> when applied to more complex systems (e.g., <i>partial</i> differential equations in space and time). We will return to this method at a later time. At the moment, it doesn't help us much.

<h3>Second-order Runge-Kutta (RK2)</h3>

There exists a family of schemes which can be collected under the umbrella of <b>Runge-Kutta</b> methods. The basic idea is based on the observation that error arises in the Euler method(s) due to replacing the right-hand side of the differential equation with $N(t)$ (or $N(t+\Delta t)$). The left-hand side <i>really</i> approximates the derivative at the point halfway between $t$ and $t + \Delta t$. So using $N(t+\Delta t/2)$ on the right hand side would be optimal. But this time is not on our "grid." Runge-Kutta methods essentially use a Taylor expansion to iteratively approximate the right-hand side in a more symmetric way than is accomplished by the Euler approach.

Let us skip over the derivation and simply outline the scheme. For simplicity let us work in terms of the actual sampled points that we'll compute. That is, for $t_{i} = i\Delta t$, $A(t_{i})\rightarrow A_{i}$. Then forward-Euler would have us compute $A_{i+1}$ from $A_{i}$ via

$$A_{i+1} = A_{i} + \Delta t (-\lambda A_{i})$$

In the second-order Runge-Kutta (RK2) approach, we write

$$A_{i+1} = A_{i} + k_{2}$$

where 

$$k_{2} = \Delta t(-\lambda (A_{i} + \frac{1}{2}k_{1}))$$

$$k_{1} = \Delta t(-\lambda A_{i})$$

The two-step process effectively approximates $A_{i+\frac{1}{2}}$ on the right-hand side. In general for a differential equation of the form

$$\frac{dx}{dt} = f(x,t)$$,

the RK2 algorithm gives $x_{i+1}$ by

$$k_{1} = \Delta t f(x_{i},t_{i})$$

$$k_{2} = \Delta t f(x_{i} + \frac{1}{2}k_{1}, t_{i} + \frac{1}{2}\Delta t)$$

$$x_{i+1} = x_{i} + k_{2}$$

Our radioactive decay example is a little simpler because there is no time dependence, but the approach works for such cases. The basic steps at each time step are: (1) compute $k_{1}$ based on $x_{i}$, $t_{i}$; (2) compute $k_{2}$ from $x_{i}$, $t_{i}$, and $k_{1}$; compute $x_{i+1}$ from $x_{i}$ and $k_{2}$. Note that $k_{1}$ is just the update step that we used in forward-Euler. By computing the $k_{2}$, we get a more symmetric approximation of the derivative.

Here's how it looks in action. I'll use `A2` as the array for $A(t)$ so we can compare to the forward-Euler results.

```python
N = 1000
tmax = 10
t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

λ = 0.5
Ao = 1000
A2 = np.zeros(N)
A2[0] = Ao

for i in range(0,N-1):
    k1 = dt*(-λ*A2[i])
    k2 = dt*(-λ*(A2[i]+0.5*k1))
    A2[i+1] = A2[i] + k2

plt.plot(t,np.abs(Ao*np.exp(-λ*t)-A))
plt.plot(t,np.abs(Ao*np.exp(-λ*t)-A2))
plt.legend(['Euler','RK2'])
plt.xlabel('t')
plt.ylabel('error in A(t)')
plt.show()
```

Quite a difference! The error of RK2 is so small that it looks like nothing. But we can take a closer look by using a logarithmic scale to stretch out the scale at smaller values:

```python
plt.semilogy(t,np.abs(Ao*np.exp(-λ*t)-A))
plt.semilogy(t,np.abs(Ao*np.exp(-λ*t)-A2))
plt.legend(['Euler','RK2'])
plt.xlabel('t')
plt.ylabel('error in A(t)')
plt.show()
```

Getting <i>no</i> error in an approximate numerical calculation would be the largest of red flags suggesting we are not computing what we <i>think</i> we're computing. So it's reassuring to see <i>some</i> error. And it's smaller by a factor of 1000 in this case. That extra step made a dramatic improvement!

<h3>Fourth-order Runge-Kutta (RK4)</h3>

The fun does not end here. One can keep iterating to get higher-order approximations that do even better. For the complicated systems we hope to model, this will often be necessary. Anyone who has computed terms in a Taylor expansion is aware that each additional term seems to require more work than all of the previous terms combined. As it happens, there is a sweet spot for RK methods at the fourth order. Higher-order methods are possible and <i>sometimes</i> necessary, but there are diminishing returns. For this reason, RK4 is one of the all-purpose methods that is well suited for a first attempt at any system of differential equations. For a general first-order differential equation,

$$\frac{dx}{dt} = f(x,t),$$

here is the formula for the update rule:

$$k_{1} = \Delta t f(x_{i},t_{i})$$

$$k_{2} = \Delta t f(x_{i}+k_{1}/2,t_{i} + \Delta t/2)$$

$$k_{3} = \Delta t f(x_{i}+k_{2}/2,t_{i} + \Delta t/2)$$

$$k_{4} = \Delta t f(x_{i}+k_{3},t_{i} + \Delta t)$$

$$x_{i+1} = x_{i} + \frac{1}{6}\left(k_{1}+2k_{2}+2k_{3}+k_{4}\right)$$

That looks more intimidating than it is. Note that the coefficients in the last expression give more weight to the interpolations in the center, between $t_{i}$ and $t_{i+1}$. In essence, $k_{1}$ approximates the slope $\frac{dx}{dt}$ at $t_{i}$, $k_{2,3}$ approximate the slope at $t_{i+\frac{1}{2}}$, and $k_{4}$ approximates the slope at $t_{i+1}$. For our simple radioactive decay model, this translates to

$$k_{1} = \Delta t(-\lambda A_{i})$$

$$k_{2} = \Delta t(-\lambda (A_{i} + k_{1}/2))$$

$$k_{3} = \Delta t(-\lambda (A_{i} + k_{2}/2))$$

$$k_{4} = \Delta t(-\lambda (A_{i} + k_{3}))$$

$$A_{i+1} = A_{i} + \frac{1}{6}\left(k_{1}+2k_{2}+2k_{3}+k_{4}\right)$$

Here's what this scheme looks like:

```python
N = 1000
tmax = 10
t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

λ = 0.5
Ao = 1000
A3 = np.zeros(N)
A3[0] = Ao

for i in range(0,N-1):
    k1 = dt*(-λ*A3[i])
    k2 = dt*(-λ*(A3[i]+0.5*k1))
    k3 = dt*(-λ*(A3[i]+0.5*k2))
    k4 = dt*(-λ*(A3[i]+k3))
    A3[i+1] = A3[i] + (1.0/6.0)*(k1+2*k2+2*k3+k4)

plt.semilogy(t,np.abs(Ao*np.exp(-λ*t)-A))
plt.semilogy(t,np.abs(Ao*np.exp(-λ*t)-A2))
plt.semilogy(t,np.abs(Ao*np.exp(-λ*t)-A3))
plt.legend(['Euler','RK2','RK4'])
plt.xlabel('t')
plt.ylabel('error in A(t)')
plt.show()
```

It's worth keeping in mind that this absolute error is not always the most informative measure of accuracy. But it's a pretty effective demonstration of the relative accuracy of each of these methods.

The <i>exact</i> solution for this one turns out to be

$$V(t) = \frac{V_{0}}{RC\left(\left(\frac{1}{RC}\right)^{2}+ \omega^{2}\right)}\left[\frac{\cos (\omega t)}{RC} + \omega \sin(\omega t) -\frac{1}{RC}e^{-\frac{t}{RC}}\right]$$

### (a) Complete the following function to use forward-Euler to return an array of V values corresponding to the approximate solution $V(t)$ returned in the vector `V`

```python
N = 1000
tmax = .1
R = 100
C = 10e-6
f = 100
ω = 2*np.pi*f
Vo = 5

t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

def forwardEuler():

    V = np.zeros(N)
#    V[0]=Vo

    for i in range(0,N-1):
        V[i+1] = V[i]+dt*(1/(R*C))*(Vo*np.cos(ω*t[i])-V[i])
#        V[i+1] = (V[i]+dt)*(Vo*(np.cos(ω*t[i]))-V[i])/(R*C)
    
#        V=V[i]*(((Vo*np.cos(ω*t))-V[i])/(R*C))
#        V[i+1] = (V[i]+dt)*(((Vo*np.cos(ω*t))-V[i])/(R*C))
#                               vint       - vc  /  RC

#        print(V[i])

#    raise NotImplementedError()
    return V
```

```python

N = 1000
tmax = .1
t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

def RK2():
    V2 = np.zeros(N)
    V2[0] = 0
    
    for i in range(0,N-1):
#        k1 = dt*( V2[i] + ( 1/(R*C) ) * ( Vo*np.cos(ω*t[i]) - V2[i] ))
        k1 = dt*( 1/(R*C) ) * ( Vo*np.cos(ω*t[i]) - V2[i] )
#        k2 = dt*((V2[i]+(.5*k1))+(1/(R*C))*(Vo*np.cos(ω*t[i]+(.5*dt))-(V2[i]+(.5*k1))))
        k2 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+(.5*dt)))-(V2[i]+(.5*k1)))
        V2[i+1] = V2[i]+k2
    
#    raise NotImplementedError()
    return V2
```

### (c) Complete the following function to use RK4 to return an array of V values corresponding to the approximate solution $V(t)$ returned to the vector `V4` . Use the same parameters/initial conditions.

```python

N = 1000
tmax = .1
t = np.linspace(0,tmax,N)
dt = t[1]-t[0]

def RK4():
    V4 = np.zeros(N)
    V4[0] = 0
    
    for i in range(0,N-1):
        k1 = dt*( 1/(R*C) ) * ( Vo*np.cos(ω*t[i]) - V4[i] )
        k2 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+(.5*dt)))-(V4[i]+(.5*k1)))
        
        k3 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+(.5*dt)))-(V4[i]+(.5*k2)))
        k4 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+    dt)) -(V4[i]+    k3))
        V4[i+1] = V4[i]+((1/6)*(k1+(2*k2)+(2*k3)+k4))
    
#    raise NotImplementedError()
    return V4
```

Lastly, there is an analytic solution to this problem (good news). But it involves Bessel function (bad news!). "What is a Bessel function?" you might well ask. The landscape of theoretical physics leads to many integrals and differential equations. Precious few of them can be evaluated/solved in terms of <b>elementary functions</b>, like $e^{x}$, $x^{n}$, $\sin x$, and so on. Nature is efficient, and many seemingly unrelated problems lead to the same governing equations. That means you could be studying fluid flow and arrive at an equation that looks the same as something from quantum mechanics if you simply swap out "density" for "probability" or "time" for "space." When this happens often enough, someone just gives the solution a name. Likely, someone from the 1800s investigated the properties of this functions and worked out an efficient way to compute it. They are called <b>special functions</b> because they're useful but not so common that the folks who make calculators decided to make a button for them (like $e^{x}$, $\sin x$, $\ln x$, etc).

The `scipy` package has a whole library of these functions. I am writing all this to give you context for the following block. We will make use of this library later, but for now I just wanted to explain what is happening. The analytic solution can be written

$$V(t) = V_{0}\ln\left[\mathcal{A}e^{-t/\tau} + I_{0}\left(\frac{V_{ac}}{V_{0}}\right) + 2\sum_{n=1}^{\infty}I_{n}\left(\frac{V_{ac}}{V_{0}}\right)\left[\frac{\cos(n(\omega t + \frac{\pi}{2})) + \omega\tau n\sin(n(\omega t + \frac{\pi}{2}))}{1+(\omega n \tau)^{2}}\right]\right]$$

Here $\mathcal{A}$ is a constant (shown below), $\tau = CV_{0}/I_{s}$ is shorthand, and $I_{n}(x)$ is a <b>modified Bessel function of the first kind</b>. Using the package `scipy.special`, this function can be accessed as $I_{n}(x) \rightarrow$ `iv(n,x)`. Unfortunately, the constant $\mathcal{A}$ is expressed as an infinite sum:

$$\mathcal{A} = 1-I_{0}\left(\frac{V_{ac}}{V_{0}}\right) - 2\sum_{n=1}^{\infty}I_{n}\left(\frac{V_{ac}}{V_{0}}\right)\frac{\cos(\pi n/2) + \omega n \tau \sin(\pi n/2)}{1+(\omega n \tau)^{2}}$$

To evaluate such a thing, one must cap the sum at some finite $n$ that is large enough that subsequent terms contribute such small corrections as to "not matter" for whatever accuracy is desired. I have taken the liberty of simply writing this thing out in the following cell (taking 1000 terms of the infinite sum). You should read this block and try to understand the broad strokes, but feel free to move on on your first attempt and just take the following array `Vt` to represent the analytic solution:

```python
#analytic solution 
Nt = 1000
import scipy.special as sp

A = 1 - sp.iv(0,Vac/Vo)
τ = C*Vo/Is

for n in range(1,Nt):
    A = A - 2*sp.iv(n,Vac/Vo)*(np.cos(n*np.pi/2) + ω*n*τ*np.sin(n*np.pi/2))/(1+(ω*n*τ)**2)
    
x = A*np.exp(-t/τ)+sp.iv(0,Vac/Vo)

for n in range(1,Nt):
    x = x + 2*sp.iv(n,Vac/Vo)*((np.cos(n*(ω*t+np.pi/2))+ω*n*τ*np.sin(n*(ω*t+np.pi/2)))/(1+(ω*n*τ)**2))
    
Vt = Vo*np.log(x)
```

Make a plot of the absolute error of each solution with respect to the analytic solution. Use a logarithmic scale on the vertical axis. 

```python
plt.semilogy(np.abs(Vt-x))
plt.show()

# raise NotImplementedError()
```

```python
def RK4int(f,N,tmax,xo):
    t = np.linspace(0,tmax,N)
    dt = t[1]-t[0]
    x = np.zeros(N)
    x[0] = xo

    for i in range(0,N-1):
        k1 = dt*f(x[i]        ,t[i]        )
        k2 = dt*f(x[i]+(.5*k1),t[i]+(.5*dt))        
        k3 = dt*f(x[i]+(.5*k2),t[i]+(.5*dt))
        k4 = dt*f(x[i]+(   k3),t[i]+(   dt))
        x[i+1] = x[i]+((1/6)*(k1+(2*k2)+(2*k3)+k4))
        
#        raise NotImplementedError()
    return x

'''
def RK4():
    for i in range(0,N-1):
        k1 = dt*( 1/(R*C) ) * ( Vo*np.cos(ω*t[i]) - V4[i] )
        k2 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+(.5*dt)))-(V4[i]+(.5*k1)))      
        k3 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+(.5*dt)))-(V4[i]+(.5*k2)))
        k4 = dt*(1/(R*C))*(Vo*np.cos(ω*(t[i]+    dt)) -(V4[i]+    k3))
        V4[i+1] = V4[i]+((1/6)*(k1+(2*k2)+(2*k3)+k4))
    
#    raise NotImplementedError()
    return V4
'''
```
